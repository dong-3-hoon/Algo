```python
import random

from pprint import pprint
'''
def is_valid(x, y):
    return 0 <= x < n and 0<= y < n
'''
n = 5

arr = [[random.randrange(1, 11) for _ in range(n)] for _ in range(n)]

#절대값의 총합
abs_sum = 0

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
#행 우선순회
for x in range(n):
    for y in range(n):
        now = arr[x][y]
        #print(f'arr[{x}][{y}]: {arr[x][y]}')
        #현재 위치에서 절댓값 차이의 합
        temp_abs = 0
        # 4방향 탐색을 하면서 이웃한 원소의 차이의 절댓값의 합
        # 주의점: 인덱스에러가 안나게 주의해야 함
        for d in range(4):
            nx = x + dx[d]
            ny = y + dy[d]
            '''if is_valid(nx, ny)'''
            if 0 <= nx < n and 0 <= ny < n:
                # 절댓값 계산
                temp_abs += abs(arr[x][y] -arr[nx][ny])
        abs_sum += temp_abs
print(f'ans: {abs_sum}')
```

```python
#quiz
#행 5 열 6 값을 임의로 할당
#2차원 배열의 모든 원소를 순회하면서 짝수인 원소의 갯수를 세는 프로그램

import random

from pprint import pprint

n, m = 5, 6

arr = [[random.randrange(1, 11) for _ in range(m)] for _ in range(n)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
row_cnt = 0
column_cnt = 0
zigzag_cnt =0
pprint(arr)
for x in range(n):
    for y in range(m):
        if arr[x][y] % 2 == 0:
            row_cnt += 1

for y in range(m):
    for x in range(n):
        if arr[x][y] % 2 == 0:
            column_cnt +=1

for x in range(n):
    for y in range(m):
        if arr[x][y + (m-1-2*y)*(x%2)] %2 == 0:
            zigzag_cnt += 1
print(row_cnt, column_cnt, zigzag_cnt)
```

```python
#한 칸 이상 탐색
di = [0, 1, 0, -1]
dj = [1, 0, -1, 0]
N = 3
P = 3
for i in range(N):
    for j in range(N):
        for k in range(4):
            for l in range(1, P+1):
                ni = i+di[k] * l
                nj = j+dj[k] * l
                if 0<=ni<N and 0<=nj<N: #인덱스 에러 방지
                    print(i, j, ni, nj)

```
